```python
import pandas as ps

data = ps.read_csv('messy_data.csv', delimiter=', ')

data['clarity'] = data['clarity'].str.upper()
data['color'] = data['color'].str.upper()
data['cut'] = data['cut'].str.upper()
data.isna().any()
```

```python
Q1 = data['price'].quantile(0.25)
Q3 = data['price'].quantile(0.75)
IQR = Q3 - Q1
condition = (Q1 - 1.5 * IQR < data.price) & (data.price < Q3 + 1.5 * IQR)
data = data[condition].sort_values('price', ascending=False)
data.isna().any()
```

```python
data['carat'] = data['carat'].interpolate()
data['x dimension'] = data['x dimension'].interpolate()
data['y dimension'] = data['y dimension'].interpolate()
data['z dimension'] = data['z dimension'].interpolate()
data['depth'] = data['depth'].interpolate()
data['table'] = data['table'].interpolate()

data.isna().any()
```

```python
data = data.dropna()
data.isna().any()
```

```python
data.to_csv(path_or_buf='cleaned.csv')
```
